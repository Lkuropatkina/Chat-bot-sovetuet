```python
from bs4 import BeautifulSoup
import requests
from selenium import webdriver
import time
```

```python
URL = 'https://market.yandex.ru/'
```

```python
response = requests.get('https://market.yandex.ru/')
response.status_code # если 200 то всё ок
```

```python
driver = webdriver.Chrome()
driver.get(URL)
#time.sleep(10) #долго грузится
```

```python
my_button = driver.find_elements_by_xpath('//*[@id="27903767-tab"]')[0] #кнопка, открывающая каталог
my_button.click()
#time.sleep(10)
```

```python
page_code = driver.page_source
```

```python
# находим подкатегории
soup = BeautifulSoup(page_code, 'html.parser')
subcategories = soup.find_all('a', class_='_2qvOOvezty SvBTI5gwNn _1B0pXQ-Qeq')
num_of_subcategories = len(subcategories)
num_of_subcategories
```

```python
categories = soup.find_all('a', class_='_241BO-CL8_')
num_of_categories = len(categories)
num_of_categories
```

```python
str_subcategories = []
for i in subcategories:
    str_subcategories.append(i.contents[0])
str_subcategories
```
